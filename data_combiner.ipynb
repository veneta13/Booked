```python
import pandas as pd
```

```python
bbe_df = pd.read_csv("data/Best_Books_Ever.csv")
bbe_df.head()
```

```python
books_df = pd.read_csv('data/Books.csv', low_memory=False)
books_df.head()
```

Let's drop the images.

```python
books_df.drop(axis=1, labels=["Image-URL-S", "Image-URL-M", "Image-URL-L"], inplace=True)
```

Inner join based on ISBN.

```python
joined_df = pd.merge(bbe_df, books_df, left_on="title", right_on="Book-Title")
joined_df.head()
```

```python
joined_df.shape
```

```python
joined_df = joined_df.drop_duplicates(subset='title', keep="first")
```

```python
joined_df
```

```python
joined_df.info()
```

```python
ratings_df = pd.read_csv('data/Ratings.csv')
```

```python
ratings_df['ISBN'] = ratings_df['ISBN'].apply(str)
```

```python
ratings_df.info()
```

```python
filtered_ratings = ratings_df.join(joined_df.set_index('ISBN'), on='ISBN', how='inner')
```

```python
filtered_ratings
```

```python
filtered_ratings.info()
```

```python
filtered_ratings = filtered_ratings.drop(columns=['isbn', 'Book-Title', 'Book-Author', 'publisher'])
```

```python
filtered_ratings
```

```python
filtered_ratings.to_csv('combined_ratings.csv', index=False)
```
